```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(-10, 10, 50)
X, Y = np.meshgrid(xx, xx)

f1 = 4
f2 = .5
p1 = .5
p2 = 1

Z = np.sin(f1 * X * np.cos(f2 * Y + p2) + p1)
gx, gy = np.gradient(Z)

plt.quiver(xx,xx, gx, gy)
```
